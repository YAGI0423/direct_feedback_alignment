# 설명글을 보고 어떤 유형의 영화인지 예측하는 프로그램을 만들어보자

```python
df_t=pd.read_csv('~/titles.csv')
df_t2=pd.read_csv('~/titles2.csv')
df_t3=pd.read_csv('~/titles3.csv')
df_t4=pd.read_csv('~/titles4.csv')
```

```python
df_t = df_t.loc[:,['description','genres']]
df_t2 = df_t2.loc[:,['description','genres']]
df_t3 = df_t3.loc[:,['description','genres']]
df_t4 = df_t4.loc[:,['description','genres']]
```

```python
df_jang.dropna(inplace=True)
```

```python
import re
```

```python
def preprocess_sentence(sentence):
    sentence = sentence.lower().strip() 
    sentence = re.sub(r"([?.!,¿])", r" \1 ", sentence) 
    sentence = re.sub(r'[" "]+', " ", sentence) 
    sentence = re.sub(r"[^a-zA-Z?.!,¿]+", " ", sentence)
    sentence = sentence.strip()
    sentence = '<start> ' + sentence + ' <end>' 
    return sentence
```

```python
corpus = []

for sentence in df_jang['description']:
   
    if len(sentence) == 0: continue
    
    
    preprocessed_sentence = preprocess_sentence(sentence)
    corpus.append(preprocessed_sentence)
        

len(corpus)
```

```python
corpus
```

```python
def preprocess_sentence2(sentence):
    sentence = sentence.lower().strip() 
    sentence = re.sub(r"([?.!,¿])", r" \1 ", sentence) 
    sentence = re.sub(r'[" "]+', " ", sentence) 
    sentence = re.sub(r"[^a-zA-Z?.!,¿]+", " ", sentence)
    sentence = re.sub(" ", "", sentence)
    sentence = sentence[:sentence.find(',')]
    sentence = sentence.strip()
    return sentence


print(preprocess_sentence("This @_is ;;;sample        sentence."))
```

```python
y_corpus = []

for sentence in df_jang['genres']:
   
    if len(sentence) == 0: continue
    
    
    preprocessed_sentence = preprocess_sentence2(sentence)
    y_corpus.append(preprocessed_sentence)
        

y_corpus
```

```python
def genre_to_idx(x):
  return a[x]
```

```python
target=df_r['y_corpus'].apply(genre_to_idx)
```

```python
import glob
import tensorflow as tf
import os, re
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
def tokenize(corpus):
    tokenizer = tf.keras.preprocessing.text.Tokenizer(
        num_words=7000, 
        filters=' ',
        oov_token="<unk>"
    )
    
    tokenizer.fit_on_texts(corpus)
    tensor = tokenizer.texts_to_sequences(corpus)   
    tensor = tf.keras.preprocessing.sequence.pad_sequences(tensor, padding='post')  
    
    print(tensor,tokenizer)
    return tensor, tokenizer

tensor, tokenizer = tokenize(df_r['corpus'])
```

```python
for idx in tokenizer.index_word:
    print(idx, ":", tokenizer.index_word[idx])

    if idx >= 10: break
```

```python
from sklearn.model_selection import train_test_split
train = tensor  
   
(enc_train, enc_val, dec_train, dec_val) = train_test_split(train,target,test_size=0.2,shuffle=True,random_state=500)
```

```python
enc_train
```

```python
dec_train
```

```python
enc_val
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy(
    from_logits=True,
    reduction='none'
)
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(40862,512))
model.add(tf.keras.layers.LSTM(1024))
model.add(tf.keras.layers.Dense(19,activation='softmax'))
model.compile(optimizer='adam',loss=loss,metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(enc_train,dec_train, epochs=2, validation_data =[enc_val,dec_val])
```
