# Credits 파일 Null 값 확인 및 처리/감독 까메오 여부 확인

```python
for i, credits in enumerate(credits_list):
    num_nan = credits["character"].isnull().sum()
    print("{0} 플랫폼의 character NaN 값의 개수 : {1}".format(platform_names[i], num_nan))

    role_director = credits.loc[credits["role"] == "DIRECTOR"]
    role_director_nan_character_num = role_director["character"].isnull().sum()
    print("{0} 플랫폼의 감독 배역 character NaN 값의 개수 : {1}".format(platform_names[i], role_director_nan_character_num))

    role_actor = credits.loc[credits["role"] == "ACTOR"]
    role_actor_nan_character_num = role_actor["character"].isnull().sum()
    print("{0} 플랫폼의 배우 배역 character NaN 값의 개수 : {1}".format(platform_names[i], role_actor_nan_character_num))
    print("두 배역 NaN 개수 총 합 : {0}".format(role_director_nan_character_num + role_actor_nan_character_num))
    print("\n")
```

- 모든 플랫폼 까메오 없음 확인

# seasons Null 값 확인 및 처리

```python
for i, titles in enumerate(titles_list):
    num_nan = titles["seasons"].isnull().sum()
    print("{0} 플랫폼의 seasons NaN 값의 개수 : {1}".format(platform_names[i], num_nan))
    
    type_movie = titles.loc[titles["type"] == "MOVIE"]
    type_movie_nan_seasons_num = type_movie["seasons"].isnull().sum()
    print("{0} 플랫폼의 영화 seasons NaN 값의 개수 : {1}".format(platform_names[i], type_movie_nan_seasons_num))

    type_show = titles.loc[titles["type"] == "SHOW"]
    type_show_nan_seasons_num = type_show["seasons"].isnull().sum()
    print("{0} 플랫폼의 드라마 seasons NaN 값의 개수 : {1}".format(platform_names[i], type_show_nan_seasons_num))
    print("\n")
```

# IMDB 관련 변수 및 TMDB 관련 변수 Null값 처리
- IMDB score가 null이면서 동시에 TMDB score도 null인 데이터 확인

```python
amazon_titles.isnull().sum()
```

```python
for i, titles in enumerate(titles_list):
    imdb_score_nan_num = titles["imdb_score"].isnull().sum()
    imdb_votes_nan_num = titles["imdb_votes"].isnull().sum()
    tmdb_popularity_nan_num = titles["tmdb_popularity"].isnull().sum()
    tmdb_score_nan_num = titles["tmdb_score"].isnull().sum()
    print("전처리 전")
    print("{0} 플랫폼의 imdb_score NaN 값의 개수 : {1}".format(platform_names[i], imdb_score_nan_num))
    print("{0} 플랫폼의 imdb_votes NaN 값의 개수 : {1}".format(platform_names[i], imdb_votes_nan_num))
    print("{0} 플랫폼의 tmdb_popularity NaN 값의 개수 : {1}".format(platform_names[i], tmdb_popularity_nan_num))
    print("{0} 플랫폼의 tmdb_score NaN 값의 개수 : {1}".format(platform_names[i], tmdb_score_nan_num))

    # imdb_score가 nan이면서 동시에 tmdb_score가 nan인 데이터 제거
    # True, False를 덧셈 연산은 아래와 같음
    # T + T -> T
    # T + F -> T
    # F + T -> T
    # F + F -> F
    # 따라서 아래와 같이 계산함.
    # ~(~T + ~T) -> ~(F) -> T
    # ~(~T + ~F) -> ~(T) -> F
    # ~(~F + ~T) -> ~(T) -> F
    # ~(~F + ~F) -> ~(T) -> F

    # imdb_score가 nan이면서 동시에 tmdb_score가 nan인 데이터
    titles_imdb_nan_tmdb_nan = titles.loc[~(~titles["imdb_score"].isnull() + ~titles["tmdb_score"].isnull())]
    # imdb_score가 nan이면서 동시에 tmdb_score가 nan이 아닌 데이터
    titles = titles.loc[~titles["imdb_score"].isnull() + ~titles["tmdb_score"].isnull()]
    print("{0} 플랫폼의 imdb_score와 tmdb_score가 동시에 NaN 값의 개수 : {1}".format(platform_names[i], titles_imdb_nan_tmdb_nan.shape[0]))

    # 전처리 이후의 값
    imdb_score_nan_num = titles["imdb_score"].isnull().sum()
    imdb_votes_nan_num = titles["imdb_votes"].isnull().sum()
    tmdb_popularity_nan_num = titles["tmdb_popularity"].isnull().sum()
    tmdb_score_nan_num = titles["tmdb_score"].isnull().sum()
    print("\nimdb_score가 nan이면서 동시에 tmdb_score가 nan인 데이터 삭제 이후")
    print("{0} 플랫폼의 imdb_score NaN 값의 개수 : {1}".format(platform_names[i], imdb_score_nan_num))
    print("{0} 플랫폼의 imdb_votes NaN 값의 개수 : {1}".format(platform_names[i], imdb_votes_nan_num))
    print("{0} 플랫폼의 tmdb_popularity NaN 값의 개수 : {1}".format(platform_names[i], tmdb_popularity_nan_num))
    print("{0} 플랫폼의 tmdb_score NaN 값의 개수 : {1}".format(platform_names[i], tmdb_score_nan_num))

    # imdb_score가 NaN인 데이터에 tmdb_score 대입
    # tmdb_score가 NaN인 데이터에 imdb_score 대입
    titles.loc[titles["imdb_score"].isnull(), "imdb_score"] = titles.loc[titles["imdb_score"].isnull(), "tmdb_score"]
    titles.loc[titles["tmdb_score"].isnull(), "tmdb_score"] = titles.loc[titles["tmdb_score"].isnull(), "imdb_score"]
    # titles["imdb_score"].fillna(value=titles.loc[titles["imdb_score"].isnull(), "tmdb_score"])
    # titles["tmdb_score"].fillna(value=titles.loc[titles["tmdb_score"].isnull(), "imdb_score"])

    # imdb_votes 값이 NaN인 값 제거
    titles = titles.loc[~titles["imdb_votes"].isnull()]

    # 전처리 이후의 값
    imdb_score_nan_num = titles["imdb_score"].isnull().sum()
    imdb_votes_nan_num = titles["imdb_votes"].isnull().sum()
    tmdb_popularity_nan_num = titles["tmdb_popularity"].isnull().sum()
    tmdb_score_nan_num = titles["tmdb_score"].isnull().sum()
    print("\nimdb_score가 nan이면서 동시에 tmdb_score가 nan인 데이터 삭제 이후")
    print("{0} 플랫폼의 imdb_score NaN 값의 개수 : {1}".format(platform_names[i], imdb_score_nan_num))
    print("{0} 플랫폼의 imdb_votes NaN 값의 개수 : {1}".format(platform_names[i], imdb_votes_nan_num))
    print("{0} 플랫폼의 tmdb_popularity NaN 값의 개수 : {1}".format(platform_names[i], tmdb_popularity_nan_num))
    print("{0} 플랫폼의 tmdb_score NaN 값의 개수 : {1}".format(platform_names[i], tmdb_score_nan_num))

    # tmdb_popularity가 NaN인 값 제거
    titles = titles.loc[~titles["tmdb_popularity"].isnull()]

    # 전처리 이후의 값
    imdb_score_nan_num = titles["imdb_score"].isnull().sum()
    imdb_votes_nan_num = titles["imdb_votes"].isnull().sum()
    tmdb_popularity_nan_num = titles["tmdb_popularity"].isnull().sum()
    tmdb_score_nan_num = titles["tmdb_score"].isnull().sum()
    print("\ntmdb_popularity가 nan인 데이터 삭제 이후")
    print("{0} 플랫폼의 imdb_score NaN 값의 개수 : {1}".format(platform_names[i], imdb_score_nan_num))
    print("{0} 플랫폼의 imdb_votes NaN 값의 개수 : {1}".format(platform_names[i], imdb_votes_nan_num))
    print("{0} 플랫폼의 tmdb_popularity NaN 값의 개수 : {1}".format(platform_names[i], tmdb_popularity_nan_num))
    print("{0} 플랫폼의 tmdb_score NaN 값의 개수 : {1}".format(platform_names[i], tmdb_score_nan_num))

    
    # csv 저장
    save_data_path = os.path.join(save_data_dir, titles_file_names[i])
    print("저장 경로 :", save_data_path)
    titles.to_csv(save_data_path, index=False)

    print("\n\n")
```
