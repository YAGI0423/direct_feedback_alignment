# 국가명 컬럼 추가

```python
def make_unique_country_list(df, platform_names):
    country_lists = []
    all_platforms_country_lists = []
    for i, platform_name in enumerate(platform_names):
        # 데이터 프레임 중 특정 플랫폼을 추출하여 국가 데이터의 유니크 값 목록 생성
        country_df = titles.loc[titles["platform"] == platform_name]
        unique_country_list = country_df["production_countries"].unique()

        country_list = []
        # 문자열 치환 후 국가 리스트 생성
        for countries in unique_country_list:
            temp_countries = re.sub('\[', "", countries)
            temp_countries = re.sub('\]', "", temp_countries)
            temp_countries = re.sub("\'", "", temp_countries)
            temp_countries = re.sub(' ', "", temp_countries)
            temp_countries = temp_countries.split(",")

            country_list.extend(temp_countries)

        # 플랫폼 별 국가명 리스트 생성
        country_lists.append(country_list)
        # 모든 플랫폼에 대한 국가명 리스트 생성
        all_platforms_country_lists.extend(country_list)

    return country_lists, all_platforms_country_lists
```

```python
country_lists, all_platforms_country_lists = make_unique_country_list(titles, platform_names)
```

- 개별 플랫폼 별 국가 종류 리스트 생성
- 모든 플랫폼을 통합한 국가 종류 리스트 생성

```python
country_column_names = set(all_platforms_country_lists)
country_column_names
```

- 데이터에 과거의 국가를 지칭하는 국가 코드들이 포함되어 있음을 확인.
- Lebanon : 레바논
- SUHH : 구소련
- UnitedStatesofAmerica : 미국

```python
# 공백 제거
country_column_names.discard("")
# 예외사항 제거
country_column_names.discard("Lebanon")
country_column_names.discard("SUHH")
country_column_names.discard("UnitedStatesofAmerica")
# 아무 국가도 없는 경우에 대한 컬럼 추가
country_column_names.add("none_country")
country_column_names
```

- 과거의 국가 코드들은 삭제하고 데이터 분리 시 예외처리 수행하겠음.

# 국가명 컬럼 데이터 추가

```python
# 국가명 컬럼 추가 후 0으로 값 설정
for country_column_name in country_column_names:
    if country_column_name == "none_country":
        continue
    titles[country_column_name] = 0
titles["none_country"] = 0
```

```python
# 데이터 프레임에서 제작 국가의 컬럼 인덱스 찾기
titles_columns = list(titles.columns)
production_countries_index = titles_columns.index("production_countries")

for i, countries in enumerate(titles["production_countries"]):
    temp_countries = re.sub('\[', "", countries)
    temp_countries = re.sub('\]', "", temp_countries)
    temp_countries = re.sub("\'", "", temp_countries)
    temp_countries = re.sub(' ', "", temp_countries)
    temp_country_list = temp_countries.split(",")

    for temp_country in temp_country_list:
        # 예외처리1(공백 즉 ''일 경우 none_country에 1 추가)
        if temp_country == "":
            production_countries_index = titles_columns.index("none_country")
            titles.iloc[i, production_countries_index] = 1
            continue

        # 예외처리2(국가명 코드가 Lebanon일 경우 LB에 1 추가)
        if temp_country == "Lebanon":
            production_countries_index = titles_columns.index("LB")
            titles.iloc[i, production_countries_index] = 1
            continue

        # 예외처리3(국가명 코드가 SUHH일 경우 SU에 1 추가)
        if temp_country == "SUHH":
            production_countries_index = titles_columns.index("SU")
            titles.iloc[i, production_countries_index] = 1
            continue

        # 예외처리4(국가명 코드가 UnitedStatesofAmerica 경우 US에 1 추가)
        if temp_country == "UnitedStatesofAmerica":
            production_countries_index = titles_columns.index("US")
            titles.iloc[i, production_countries_index] = 1
            continue

        # 국가명 컬럼에 데이터 추가
        production_countries_index = titles_columns.index(temp_country)
        titles.iloc[i, production_countries_index] = 1
```

# 장르명 컬럼 추가

```python
def make_unique_genre_list(df, platform_names):
    genre_lists = []
    all_platforms_genre_lists = []
    for i, platform_name in enumerate(platform_names):
        # 데이터 프레임 중 특정 플랫폼을 추출하여 국가 데이터의 유니크 값 목록 생성
        genre_df = titles.loc[titles["platform"] == platform_name]
        unique_genre_list = genre_df["genres"].unique()

        genre_list = []
        # 문자열 치환 후 국가 리스트 생성
        for genres in unique_genre_list:
            temp_genres = re.sub('\[', "", genres)
            temp_genres = re.sub('\]', "", temp_genres)
            temp_genres = re.sub("\'", "", temp_genres)
            temp_genres = re.sub(' ', "", temp_genres)
            temp_genres = temp_genres.split(",")

            genre_list.extend(temp_genres)

        # 개별 플랫폼에 대한 장르명 리스트 생성
        genre_lists.append(genre_list)
        # 모든 플랫폼에 대한 장르명 리스트 생성
        all_platforms_genre_lists.extend(genre_list)

    return genre_lists, all_platforms_genre_lists
```

```python
genre_lists, all_platforms_genre_lists = make_unique_genre_list(titles, platform_names)
```

```python
genre_column_names = set(all_platforms_genre_lists)
genre_column_names
```

# 장르명 컬럼 데이터 추가

```python
# 장르명 컬럼 추가 후 0으로 값 설정
for genre_column_name in genre_column_names:
    if genre_column_name == "none_genre":
        continue
    titles[genre_column_name] = 0
titles["none_genre"] = 0
```

```python
# 데이터 프레임에서 장르의 컬럼 인덱스 찾기
titles_columns = list(titles.columns)
genres_index = titles_columns.index("genres")

for i, genres in enumerate(titles["genres"]):
    temp_genres = re.sub('\[', "", genres)
    temp_genres = re.sub('\]', "", temp_genres)
    temp_genres = re.sub("\'", "", temp_genres)
    temp_genres = re.sub(' ', "", temp_genres)
    temp_genres_list = temp_genres.split(",")

    for temp_genre in temp_genres_list:
        # 예외처리1(공백 즉 ''일 경우 none_genre에 1 추가)
        if temp_genre == "":
            genres_index = titles_columns.index("none_genre")
            titles.iloc[i, genres_index] = 1
            continue

        # 장르명 컬럼에 데이터 추가
        genres_index = titles_columns.index(temp_genre)
        titles.iloc[i, genres_index] = 1
```

# 제한 연령 컬럼 및 데이터 추가

```python
titles["age_certification"].unique()
```

- 미국 기준의 연령 제한만 존재함.
- all : G, TV-G, TV-Y
- 7+ : PG, TV-Y7, TV-PG
- 13+ : PG-13
- 16+ : TV-14
- 18+ : NC-17, R, TV-MA

```python
titles["num_age_certification"] = "none"
```

```python
# 데이터 프레임에서 age_certification의 컬럼 인덱스 찾기
titles_columns = list(titles.columns)
age_column_index = titles_columns.index("num_age_certification")

for i, age in enumerate(titles["age_certification"]):
    # all
    if age == "G" or age == "TV-G" or age == "TV-Y":
        titles.iloc[i, age_column_index] = "all"
    # 7+
    elif age == "PG" or age == "TV-Y7" or age == "TV-PG":
        titles.iloc[i, age_column_index] = "7+"
    # 13+
    elif age == "PG-13":
        titles.iloc[i, age_column_index] = "13+"
    # 16+
    elif age == "TV-14":
        titles.iloc[i, age_column_index] = "16+"
    # 18+
    elif age == "NC-17" or age == "R" or age == "TV-MA":
        titles.iloc[i, age_column_index] = "18+"
```
