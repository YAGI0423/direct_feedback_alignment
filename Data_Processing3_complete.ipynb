```python
data_dir = '/gdrive/MyDrive/0. 인공지능 학습/아이펠 대전 3기/5. 해커톤/1차 해커톤/processed_data2'
save_data_dir = '/gdrive/MyDrive/0. 인공지능 학습/아이펠 대전 3기/5. 해커톤/1차 해커톤/processed_data3'
credits_file_names = ["amazon_credits.csv", "netflix_credits.csv", "disney_credits.csv", "paramount_credits.csv"]
titles_file_names = ["amazon_titles.csv", "netflix_titles.csv", "disney_titles.csv", "paramount_titles.csv"]
platform_names = ["amazon", "netflix", "disney", "paramount"]

# 아마존
amazon_credits_path = os.path.join(data_dir, credits_file_names[0])
amazon_titles_path = os.path.join(data_dir, titles_file_names[0])
# 넷플릭스
netflix_credits_path = os.path.join(data_dir, credits_file_names[1])
netflix_titles_path = os.path.join(data_dir, titles_file_names[1])
# 디즈니
disney_credits_path = os.path.join(data_dir, credits_file_names[2])
disney_titles_path = os.path.join(data_dir, titles_file_names[2])
# 파라마운트
paramount_credits_path = os.path.join(data_dir, credits_file_names[3])
paramount_titles_path = os.path.join(data_dir, titles_file_names[3])
```

```python
# 아마존
amazon_credits = pd.read_csv(amazon_credits_path)
amazon_titles = pd.read_csv(amazon_titles_path)
# 넷플릭스
netflix_credits = pd.read_csv(netflix_credits_path)
netflix_titles = pd.read_csv(netflix_titles_path)
# 디즈니
disney_credits = pd.read_csv(disney_credits_path)
disney_titles = pd.read_csv(disney_titles_path)
# 파라마운트
paramount_credits = pd.read_csv(paramount_credits_path)
paramount_titles = pd.read_csv(paramount_titles_path)
```

```python
credits_list = [amazon_credits, netflix_credits, disney_credits, paramount_credits]
titles_list = [amazon_titles, netflix_titles, disney_titles, paramount_titles]
```

# credits 데이터 프레임 병합 및 회사명 추가

```python
# 데이터 병합 전 credits 파일과 동일한 columns 명을 갖는 빈 데이터 프레임 생성
blank_df = pd.DataFrame(data=[], columns=amazon_credits.columns)
# blank_df
```

```python
ex_df = blank_df
for i, credits in enumerate(credits_list):
    # 회사명 컬럼 추가 및 회사명 데이터 추가
    credits["platform"] = platform_names[i]
    temp_df = credits
    # 데이터 프레임 병합
    ex_df = pd.concat([ex_df, temp_df], axis=0)

# csv 저장
save_data_path = os.path.join(save_data_dir, "merged_credits.csv")
print("저장 경로 :", save_data_path)
ex_df.to_csv(save_data_path, index=False)
```

# titles 데이터 프레임 병합 및 회사명 추가

```python
# 데이터 병합 전 titles 파일과 동일한 columns 명을 갖는 빈 데이터 프레임 생성
blank_df = pd.DataFrame(data=[], columns=amazon_titles.columns)
# blank_df
```

```python
ex_df = blank_df
for i, titles in enumerate(titles_list):
    # 회사명 컬럼 추가 및 회사명 데이터 추가
    titles["platform"] = platform_names[i]
    temp_df = titles
    # 데이터 프레임 병합
    ex_df = pd.concat([ex_df, temp_df], axis=0)
    
# csv 저장
save_data_path = os.path.join(save_data_dir, "merged_titles.csv")
print("저장 경로 :", save_data_path)
ex_df.to_csv(save_data_path, index=False)
```
